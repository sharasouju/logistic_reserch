```python
# 必要なライブラリとデータを読み込む
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
data = pd.read_csv("ptitanic.csv")
```

```python
#データ
data.head()
```

```python
data.describe()
```

```python
#欠損値
data_ti = data.dropna()
```

```python
# サンプルサイズを確認
len(data), len(data_ti)
```

```python
# 性別ごとの乗船者数
%matplotlib inline
sns.countplot(x='sex', data=data_ti)
sns.plt.show()
```

```python
# 年齢ごとの乗船者数
sns.distplot(data_ti['age'])
sns.plt.show()
```

```python

grid = sns.FacetGrid(data_ti, col="sex", hue="sex")
grid.map(sns.countplot, 'age')
```

```python
# 年代ごとのラベルを設定
data_ti.loc[:,'agecut']=pd.cut(data_ti['age'], labels=["Baby", "Child", "Youth", "Adult", "old"], bins=[float("-Inf"), 1, 12, 18, 58, float("Inf")])
```

```python
# 性別 x 年代ごとの乗船者数
grid = sns.FacetGrid(data_ti, col="sex", hue="sex")
grid.map(sns.countplot, 'agecut')
```

```python
# ---------------------------
# 分類木の実践
# ---------------------------
import pandas
import sklearn.model_selection
train, test = sklearn.model_selection.train_test_split(data_ti, train_size = 0.8)
```

```python
from sklearn.preprocessing import LabelEncoder
train_encoded=train.apply(LabelEncoder().fit_transform)
test_encoded=test.apply(LabelEncoder().fit_transform)
```

```python
train_encoded.isnull().sum(axis=0)
```

```python
# 学習
from sklearn.tree import DecisionTreeClassifier
X=train_encoded.loc[ : , ['pclass','sex','agecut']]
y=train_encoded['survived']
dt = DecisionTreeClassifier()
dt.fit(X, y)
```

```python
# 学習結果をプロット
from sklearn import tree
from IPython.display import Image
import pydotplus
dot_data = tree.export_graphviz(dt, out_file=None, 
                         feature_names=['pclass','sex','agecut'],  
                         class_names='survived',  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  
```

```python
# 予測
pred = dt.predict(test_encoded.loc[:, ['pclass','sex','agecut']])
```

```python
# 予測の結果についてconfusion matrixを作成
import pandas as pd
import seaborn as sn
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

def print_cmx(y_true, y_pred):
    labels = sorted(list(set(y_true)))
    cmx_data = confusion_matrix(y_true, y_pred, labels=labels)
    
    df_cmx = pd.DataFrame(cmx_data, index=labels, columns=labels)

    plt.figure(figsize = (10,7))
    sn.heatmap(df_cmx, annot=True)
    plt.show()
```

```python
print_cmx(test_encoded.survived, pred)
```

```python
# 正解率
from sklearn.metrics import accuracy_score
accuracy_score(test_encoded.survived, pred)
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston_data = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
#データ
boston_data.head()
```

```python
boston_data.describe()
```

```python
boston_data.isnull().sum(axis=0)
```

```python
%matplotlib inline
sns.distplot(boston.target, kde=False, rug=False, bins=25)
sns.plt.show()
```

```python
boston_data_2=pd.concat([boston_data, pd.DataFrame(boston.target,columns=['MEDV'])], axis=1)
sns.jointplot("CRIM", "MEDV", data=boston_data_2)
```

```python
sns.jointplot("LSTAT", "MEDV", data=boston_data_2)
```

```python
# 住居の平均部屋数
sns.jointplot("RM", "MEDV", data=boston_data_2)
```

```python
# 1940 年より前に建てられた物件の割合
sns.jointplot("AGE", "MEDV", data=boston_data_2)
```

```python
# 訓練データとテストデータに分割
import pandas
import sklearn.model_selection
boston_train, boston_test = sklearn.model_selection.train_test_split(boston_data_2, train_size = 0.8)
```

```python
#学習
from sklearn.tree import DecisionTreeClassifier
boston_X=boston_train.loc[:, ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
boston_y=pd.DataFrame(boston_train['MEDV'].astype('int')).loc[:,['MEDV']]
boston_dt = DecisionTreeClassifier()
boston_dt.fit(boston_X, boston_y)
```

```python
boston_X
```

```python
# 変数重要度
importances = boston_dt.feature_importances_
indices = np.argsort(importances)[::-1]
print("Feature ranking:")

for f in range(boston_X.shape[1]):
    print("%d. %s (%f)" % (f + 1, boston_X.columns[indices[f]], importances[indices[f]]))
```

```python
#予測
boston_test_X=boston_test.loc[:, ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
boston_test_predict = boston_dt.predict(boston_test_X)
```

```python
# 予測の誤差をプロット
boston_res = boston_test_predict - boston_test.MEDV
sns.distplot(boston_res, kde=False)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# 標準誤差
rms = sqrt(mean_squared_error(boston_test.MEDV, boston_test_predict))
rms
```

```python
# 平均絶対誤差（mean absolute error）
from sklearn.metrics import mean_absolute_error
mean_absolute_error(boston_test.MEDV, boston_test_predict)
```

```python
from rpy2.robjects import r, pandas2ri
def data(name): 
    return pandas2ri.ri2py(r[name])
```

```python
mtcars_data = data('mtcars')
```

```python
mtcars_data.head()
```

```python
mtcars_data.describe()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.lmplot(x='wt', y='hp', hue='am', data=mtcars_data, fit_reg=False)
```

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
X = mtcars_data.loc[:, ['hp','wt']]
y = mtcars_data['am']
clf = LogisticRegression()

clf.fit(X, y)
```

```python
import statsmodels.api as sm

logit_model = sm.Logit(y, sm.add_constant(X)).fit()
print logit_model.summary()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
# グラフ作成
fig = plt.figure()
ax = Axes3D(fig)

# 軸ラベルの設定
ax.set_xlabel("weight")
ax.set_ylabel("horse power")
ax.set_zlabel("probability of automatic")

x = np.linspace(mtcars_data.wt.min(), mtcars_data.wt.max(), 50)
y = np.linspace(mtcars_data.hp.min(), mtcars_data.hp.max(), 50)

data = pd.DataFrame([(i, j) for i in x for j in y])
data.columns = ['wt','hp']

X, Y = np.meshgrid(x, y)
z = clf.predict_proba(data)[:,1]
Z = z.reshape(50,50)

ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())
ax.set_zlim(z.min(), z.max())

ax.plot_surface(X,Y,Z,cmap=cm.coolwarm) 

plt.show()
```
